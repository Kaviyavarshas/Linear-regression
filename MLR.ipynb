# Detecting influencers/outliers

```python
#1.Cooks distance=inbuilt fxn in stats_models.directly influence is got &can cal dist
#2.high influ pts
#3.Cooks dist is used in REgression analysis to find influential outliers in a set of ID variables.

#in othr wrds,it is a way to identify pts that negatively affect ur regressin mod
#measurement is a combo of each obs's leverage & residual vals.
#hihger the lev & residues,the hhigher the cook's dist
#several interpretation fr cook's dist exist.thr isn't a universally accepted rule fr cut off pts.
#if thr are lots of pts wid large cook"s dist vals,that could indicate a pblm wid regression mod in general.
```

## 1.Cooks distance

```python
model_f2_Outliers=model_f2.get_influence()
```

```python
dis,_=model_f2_Outliers.cooks_distance
```
